```python
import os

import numpy as np
import tensorflow as tf

import config, music_model_vectorized

ROOT_PATH = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
conf = config.Config("single_instruments_type", ROOT_PATH)

batch_size = 2
```

```python
model = music_model_vectorized.create_debugging_model(
    decoder_output_shape=[6144, 512],
    song_shape=[6143, 11]
)

model.compile(optimizer="adam")
```

```python
dataset = tf.data.Dataset.load(conf.tf_data7dict_path)
dataset = dataset.batch(batch_size)
```

```python
for item in dataset.take(1):
    songs = item[0][0]
```

```python
scores = np.random.random((batch_size, 6144, 512))
```

```python
def print_stuff(inputs):
    a, b = inputs
    tf.print(a, "-", b)
    print(a, "-", b)
    return a

mask = tf.vectorized_map(
            fn=print_stuff,
            elems=(
                tf.range(10), 
                tf.expand_dims([0, 1, 5, 7], axis=0)
            )
        )

print(mask)
```

```python
model((scores, songs))
```
